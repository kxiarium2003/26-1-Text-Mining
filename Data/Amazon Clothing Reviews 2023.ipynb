```python
from huggingface_hub import list_repo_files

repo_id = "McAuley-Lab/Amazon-Reviews-2023"
files = list_repo_files(repo_id, repo_type="dataset")

fashion_files = [f for f in files if "Clothing" in f or "Fashion" in f]


for f in fashion_files:
    print(f)
```

```python
import pandas as pd


base_url = "https://huggingface.co/datasets/McAuley-Lab/Amazon-Reviews-2023/resolve/main/"
file_path = "raw/review_categories/Amazon_Fashion.jsonl"
url = base_url + file_path

try:
    df_sample = pd.read_json(url, lines=True, nrows=100)

    print("데이터 로드 완료.")
    display(df_sample[['rating', 'title', 'text']].head(10))

    print("\n[상세 리뷰 샘플 (상위 10개)]")
    print(df_sample['text'].iloc[10])

except Exception as e:
    print(f"로드 실패 (에러 메시지 확인): {e}")
```

```python
# 평점이 2점 이하인 리뷰만 추출
low_rating_df = df_sample[df_sample['rating'] <= 2]

print(f"필터링된 리뷰 개수: {len(low_rating_df)}")
display(low_rating_df[['rating', 'text']].head(10))

# 이 리뷰들에서 가장 많이 나온 단어 확인
from collections import Counter
all_text = " ".join(low_rating_df['text'].tolist()).lower()
words = all_text.split()
common_words = Counter(words).most_common(20)
print("\n[부정 리뷰 주요 키워드]\n", common_words)
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))
filtered_words = [w for w in words if not w in stop_words and len(w) > 1]
```
