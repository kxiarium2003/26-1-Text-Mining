```python
import pandas as pd


base_url = "https://huggingface.co/datasets/McAuley-Lab/Amazon-Reviews-2023/resolve/main/"
file_path = "raw/review_categories/Amazon_Fashion.jsonl"
url = base_url + file_path

try:
    df_sample = pd.read_json(url, lines=True, nrows=100)

    print("데이터 로드 완료.")
    display(df_sample[['rating', 'title', 'text']].head(10))

    print("\n[상세 리뷰 샘플 (상위 10개)]")
    print(df_sample['text'].iloc[10])

except Exception as e:
    print(f"로드 실패 (에러 메시지 확인): {e}")
```

```python
# 평점이 2점 이하인 리뷰만 추출
low_rating_df = df_sample[df_sample['rating'] <= 2]

print(f"필터링된 리뷰 개수: {len(low_rating_df)}")
display(low_rating_df[['rating', 'text']].head(10))

# 이 리뷰들에서 가장 많이 나온 단어 확인
from collections import Counter
all_text = " ".join(low_rating_df['text'].tolist()).lower()
words = all_text.split()
common_words = Counter(words).most_common(20)
print("\n[부정 리뷰 주요 키워드]\n", common_words)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 1. 데이터 로드 설정
base_url = "https://huggingface.co/datasets/McAuley-Lab/Amazon-Reviews-2023/resolve/main/"
file_path = "raw/review_categories/Amazon_Fashion.jsonl"
url = base_url + file_path

try:
    df_sample = pd.read_json(url, lines=True, nrows=5000)

    # 2. 데이터프레임 구조 및 기본 정보 출력
    print("="*50)
    print(f"데이터 로드 완료. 전체 크기: {df_sample.shape}")
    print(f"컬럼명: {df_sample.columns.tolist()}")
    print("-" * 50)
    print(df_sample.info())

    # 3. 결측치 확인
    print("\n[결측치 확인]")
    print(df_sample.isnull().sum())

    # 4. Random Seed 고정 및 무작위 샘플링
    np.random.seed(42)
    sample_check = df_sample.sample(n=10, random_state=42)

    print("\n[무작위 추출된 샘플 리뷰 (상위 5개)]")
    display(sample_check[['rating', 'title', 'text']].head())

    # 5. 평점 분포 시각화
    plt.figure(figsize=(8, 5))
    sns.set_style("whitegrid")
    # countplot 생성
    ax = sns.countplot(x='rating', data=df_sample, palette='viridis')

    plt.title('Distribution of Ratings (Amazon Fashion Sample)', fontsize=15)
    plt.xlabel('Rating (1 to 5)', fontsize=12)
    plt.ylabel('Count', fontsize=12)

    # 막대 위에 빈도수 표시
    for p in ax.patches:
        ax.annotate(f'{int(p.get_height())}', (p.get_x()+0.3, p.get_height()+10))

    plt.show()

except Exception as e:
    print(f"로드 실패 (에러 메시지 확인): {e}")
```
